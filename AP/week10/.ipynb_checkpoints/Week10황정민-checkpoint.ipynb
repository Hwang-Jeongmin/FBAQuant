```python
import numpy as np
import pandas as pd

import FinanceDataReader as fdr

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

import warnings
warnings.filterwarnings('ignore')
```

```python
stock_df = fdr.StockListing('S&P500')
tickers = stock_df['Symbol'].tolist()
```

```python
print(f"Number of Stocks: {len(tickers)}")
```

```python
start_date = "2020-01-01"
end_date = "2023-05-11"
```

```python
df = pd.DataFrame()
for ticker in tickers:
    try:
        temp = fdr.DataReader(ticker, start_date, end_date)
        df[ticker] = temp['Adj Close']
    except:
        pass
    
df.to_csv("sp500_df.csv")
```

```python
df = pd.read_csv("sp500_df.csv", index_col=0)
df = df.dropna(axis=1)
```

```python
returns = df.pct_change().dropna()

returns.index = pd.to_datetime(returns.index, format='%Y-%m-%d')
cols = returns.columns.tolist()
idx = returns.index.tolist()

print(f"Number of Stocks: {len(df.columns)}")
```

```python
scaler = StandardScaler()
scaled_returns = scaler.fit_transform(returns)
scaled_returns = pd.DataFrame(returns, columns=cols, index=idx)
```

```python
num = 10
pca = PCA(n_components=num)
PrincipalComponent = pca.fit(scaled_returns)
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(14,4))
Series1 = pd.Series(pca.explained_variance_ratio_[:num]).sort_values(ascending=True)
Series2 = pd.Series(pca.explained_variance_ratio_[:num]).sort_values(ascending=False).cumsum()
Series1.plot.barh(title='Explained Variance Ratio by Top Factors', ax=axes[0])
Series2.plot(ylim=(0,1), ax=axes[1], title='Cumulative Explained Variance')
plt.show()
```

```python
first_component = pca.components_[0] / sum(pca.components_[0])
first_component[:10]
```

```python
def calculate_cumulative_return(daily_return):
    return np.exp(np.log(daily_return+1).cumsum())-1
pca_returns = returns.mul(first_component, axis=1).sum(axis=1)
pca_cumulative_return = calculate_cumulative_return(pca_returns)
```

```python
sp500 = fdr.DataReader("US500", start_date, end_date)
sp500 = sp500['Adj Close']
sp500_returns = sp500.pct_change().dropna()
sp500_cumulative_return = calculate_cumulative_return(sp500_returns)
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
df= pd.DataFrame({'PCA': pca_cumulative_return, 'SP500': sp500_cumulative_return})
ax.plot(df, label=['PCA', 'SP500'])
ax.set_title('CUMULATIVE RETURN', fontsize=16)
ax.legend(loc='best', fontsize=12)
ax.axhline(y=0, color='black', linestyle='dashed')
ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
plt.show()
```
