### CH1  




Basic Pricing Equation  
$ p_t = E_t[\beta \frac{u'(c_{t+1})}{u'(c_t)}x_{t+1})] $  
($ x_{t+1} $ = payoff of t+1,  $ x_{t+1} = p_{t+1} + d_{t+1} $)   
$ U(c_t, c_{t+1}) = u(c_t) + \beta E_t[u(c_{t+1})] $  
($ c_t$ consumption at date t, $u(c_t) = \frac{1}{1-\gamma}c_t^{1-\gamma} $)    
$ \\ $  
we can break up basic consumption-based pricing equation into  
$ p = E[mx], m = \beta\frac{u'(c_{t+1})}{u'(c_t)} $  
we call m as stochastic discount factor, marginal rate ofsubstitution, pricing kernel, etc.  

THe risk-free rate is related to the discount factor by  
$R^f = 1/E(m)$  
Real interest rates are high when people are impatient, when expected growth is high, or risk is low.  A more curved utility function or a lower elasticity of intertemporal substitution means that interest rates are more sensitive to changes in expected consumption growth.  

payoffs that are positively correlated with consumption growth have lower prices, to compensate investors for risk.  
$ p = \frac{E(x)}{R^f} + cov(m,x) $   
$ E(r^i)-R^f = -R^fcov(m, r^i) $  
(cov(m,x) = E(mx) - E(m)E(x))   
$ \\ $  
Only the component of a payoff perfectly correlated with the discount factor generate an extra return.Idiosyncratic risk, uncorrelated with the discount factor, generates no premium.  
$ \\ $  
We can write $ p = E(mx) $ as  
$ E(R^i) = R^f + \beta_{i,m}\lambda_m $  (beta-pricing model)   




All asset returns lie insdie mean-variance frontier.Assets on the frontier are perfectly correlated with each other and with discount factor. Returns on the frontier can be geneated as portfolios of any two frontier returns.   
$ \\ $  
The Sharpe ratio is limited by the volatility of the distcount factor. The maximal risk-return trade-ff is steeper if there is more risk or more risk- aversion.  
$ |\frac{E(R) - R^f}{\sigma(R)}| \le \frac{\sigma(m)}{E(m)} $  
If investors are risk neutral, returns are unpredictable, and prices follow martiangles.  
$ \\ $  
Present-Value Statement  
$ p_t = E_t[\Sigma_{j=1}^{\infty}m_{t,t+j}d_{t+j}] $   
$ p_t = E_t[m_{t+1}(p_{t+1} + d_{t+1})] $
